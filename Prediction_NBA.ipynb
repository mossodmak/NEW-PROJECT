```
from keras.models import load_model
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

model = load_model('drive/My Drive/model_NBA.h5')
img = cv2.imread('drive/My Drive/Machine_NBA/train/Chicago_Bulls/3.jpg')
img = cv2.resize(img,(128,128))
img = np.reshape(img,[1,128,128,3])
pred = model.predict(img)
img_path = 'drive/My Drive/Machine_NBA/train/Chicago_Bulls/3.jpg'

image = mpimg.imread(img_path)
plt.imshow(image)
plt.show()
if pred.argmax() == 0:
  print('Boston_Celtics_0')
if pred.argmax() == 1:
  print('Brooklyn_Nets_1')
if pred.argmax() == 2:
  print('Philadelphia_76ers_2')  
if pred.argmax() == 3:
  print('Cleveland_Cavaliers_3')  
if pred.argmax() == 4:
  print('Detroit_Pistons_4')  
if pred.argmax() == 5:
  print('Indiana_Pacers_5')  
if pred.argmax() == 6:
  print('Milwaukee_Bucks_6')  
if pred.argmax() == 7:
  print('NewYork_Knicks_7')  
if pred.argmax() == 8:
  print('Chicago_Bulls_8')  
if pred.argmax() == 9:
  print('Toronto_Raptors_9')  
```
